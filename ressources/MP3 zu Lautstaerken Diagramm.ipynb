```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import librosa as lib
```

```python
def mp3_to_dB_series(path):
    signal, sr = lib.load(path)
    return lib.amplitude_to_db(np.abs(signal), ref=np.max(signal)), sr
```

```python
def mp3_to_dB_diagram(path, window):
    y = mp3_to_dB_series(path)[0]
    y_avg = []
    
    for ind in range(len(y) - window + 1):
        y_avg.append(np.mean(y[ind:ind+window]))
        pass

    figure(figsize=(15, 6), dpi=200)
    
    plt.plot(y, linewidth=0.05, color='lightskyblue')
    plt.plot(y_avg, linewidth=1.5, color='darkblue', label='Moving Avg. with Window '+str(window))
    plt.show()
    pass
```
