```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import librosa as lib
```

```python
def plot_spectogram(y, sr, hop_size, y_axis='linear'):
    plt.figure(figsize = (25,10))
    lib.display.specshow(y,
                             sr=sr,
                             hop_length=hop_size,
                             x_axis="time",
                             y_axis=y_axis,
                        fmin= np.min(y),
                        fmax= np.max(y))
    plt.colorbar(format="%+2.f")
```

```python
def mp3_to_spectogram(path):
    audio_file, sr = lib.load(path)
    frame_size = 2048
    hop_size = 512
    
    s_audio_file = lib.stft(audio_file, n_fft=frame_size, hop_length=hop_size)
    y_audio_file = np.abs(s_audio_file)**2
    y_log_audio_file = lib.power_to_db(y_audio_file)
    
    plot_spectogram(y_log_audio_file,sr,hop_size,y_axis='log')
    
    pass
```
