```python
filler_words = []
```

```python
def filler_labeling_for_dataframe(data):
    df = copy.deepcopy(data)
    labels = []
    
    for i in range(len(df[0])):
        if df[0][i] in filler_words:
            labels.append(1)
            pass
        else:
            labels.append(0)
            pass
        pass
    
    df.append(labels)
    
    return df
```

```python
def moving_average_diagramm(y, window):
    y_avg = []
    
    for ind in range(len(y) - window + 1):
        y_avg.append(np.mean(y[ind:ind+window]))
        pass
    
    figure(figsize=(15, 6), dpi=200)
    
    plt.plot(y_avg, linewidth=1.5, color='darkblue', label='Moving Avg. with Window '+str(window))
    plt.show()
    pass
```

```python
y = [np.random.randint(0,2) for i in range(1000)]
```

```python
moving_average_diagramm(y,250)
```
